```python
import stasm
import cv2
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
img = cv2.imread('/content/drive/MyDrive/face.jpg',cv2.IMREAD_GRAYSCALE)
landmarks  = stasm.search_single(img)
```

```python
landmarks  = stasm.search_single(img)
```

```python
print(len(landmarks))
```

```python
fig = plt.figure(figsize = (8,8))
ax =fig.add_subplot(1,2,1)
ax.imshow(img,cmap = 'gray')
```

```python
for pt in landmarks:
  cv2.circle(img,(pt[0],pt[1]),5,(0,0,255),-1)

g = plt.figure(figsize = (10,10))
ax2 = g.add_subplot(1,2,2)
ax2.imshow(img,cmap ='gray')
```
