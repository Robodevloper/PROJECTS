```python
 def con_num(x):
        token = x.split('-')
        
        if len(token) == 2:
            return (float(token[0])+float(token[1]))/2
            
        try:
            return float(x)
         
        except:
             
            return None
                    
                    
        
```

```python
con_num('1133 - 1384')
```

```python
df4 = df3.copy()
```

```python
df4['total_sqft'] = df4['total_sqft'].apply(con_num)
```

```python
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
```

```python
def rmv_price_outlier(df):
    df_new=pd.DataFrame()
    for key,sdf in df.groupby('location'):
        m = sdf['price_per_sqft'].mean()
        s = sdf['price_per_sqft'].std()
        #print (sdf['location'])
        rdf = sdf[(sdf['price_per_sqft'] <= m+s) & (sdf['price_per_sqft'] > m-s)]
        #print(rdf)
        df_new = pd.concat([df_new,rdf],ignore_index=True)
    return df_new
```

```python
df7 = rmv_price_outlier(df6)
```

```python
import numpy as np
def predict_price(location ,sqft ,bath , bhk):
    
    loc_index = np.where(x.columns == location)[0][0]
    
    z = np.zeros(len(x.columns))
    
    z[0] = sqft
    z[1] = bath
    z[2] = bhk
    
    if loc_index>=0:
        z[loc_index] = 1
    
    return model_lr.predict([z])[0]
```
