```python
import face_recognition
import PIL.Image
import PIL.ImageDraw
```

```python
img = face_recognition.load_image_file('/content/drive/MyDrive/face.jpg')
```

```python
face_landmarks = face_recognition.face_landmarks(img)
```

```python
print(face_landmarks)
```

```python
left_eye_landmarks =  face_landmarks[0].get('left_eye')
right_eye_landmarks = face_landmarks[0].get('right_eye')
nose_bridge_landmarks = face_landmarks[0].get('nose_bridge')
bottom_lip_landmarks =  face_landmarks[0].get('bottom_lip')
```

```python
from matplotlib import pyplot as plt 
%matplotlib inline
import cv2
```

```python
for pt in left_eye_landmarks:
   
   cv2.circle(img,(pt[0],pt[1]),10,(0,0,255),-1)


for po in right_eye_landmarks:
   
   cv2.circle(img,(po[0],po[1]),10,(0,0,255),-1)   



for j in nose_bridge_landmarks:
   
   cv2.circle(img,(j[0],j[1]),10,(0,0,255),-1)




for k in bottom_lip_landmarks:
   
   cv2.circle(img,(j[0],j[1]),10,(0,0,255),-1)








```

```python
g = plt.figure(figsize = (10,10))
ax2 = g.add_subplot(1,2,2)
ax2.imshow(img,cmap ='gray')
```
